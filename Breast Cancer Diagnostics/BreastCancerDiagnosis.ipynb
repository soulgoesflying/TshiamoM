# ðŸ” 4. Above-Average Tumor Analysis

```python
average_radius = df['radius_mean'].mean()

benign = df[df['diagnosis'] == 'B']

b_above_avg = benign[benign['radius_mean'] > average_radius]

bperc_above_avg = (len(b_above_avg)/len(benign)) * 100

print(f"{bperc_above_avg:.2f}% of benign tumors were above-average radius.")
```

```python
average_radius = df['radius_mean'].mean()

malignant = df[df['diagnosis'] == 'M']

m_above_avg = malignant[malignant['radius_mean'] > average_radius]

mperc_above_avg = (len(m_above_avg)/len(malignant)) * 100

print(f"{mperc_above_avg:.2f}% of malignant tumors were above-average radius.")
```

# ðŸ”¥ 5. Correlation Heatmap

```python
top_features = ['radius_mean', 'perimeter_mean', 'area_mean',
                'concavity_mean', 'concave points_mean',
                'radius_worst', 'perimeter_worst', 'area_worst',
                'compactness_worst', 'concavity_worst']

corr_subset = above_avg_df[top_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_subset, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation of Key Features")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

# ðŸ§ª 6. Statistical Testing 

```python
from scipy.stats import ttest_ind

mal = df[df['diagnosis'] == 'M']['radius_mean']
ben = df[df['diagnosis'] == 'B']['radius_mean']
t_stat, p = ttest_ind(mal, ben)

print(f"T-stat: {t_stat:.2f}, p-value: {p:.5f}")
```
